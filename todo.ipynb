```python
from rdkit import Chem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt
from mpl_rdkit_utils import DrawMolToMPL
```

```python
smi = 'C1=CN(C(=O)N=C1N)[C@H]2[C@@H]([C@@H]([C@H](O2)' + \
      '/C=C/P(=O)(O)OP(=O)(O)OP(=O)(O)O)O)F'
mol = Chem.MolFromSmiles(smi)
```

```python
Draw.MolToImage(mol)
```

```python
Draw.MolsToGridImage([mol], useSVG=True)
```

```python
Draw.MolToMPL(mol);
```

```python
fig, ax = plt.subplots()
lim = DrawMolToMPL(mol, fig, ax)
```
