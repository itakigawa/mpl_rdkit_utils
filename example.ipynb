```python
import matplotlib.pyplot as plt
import pubchempy as pcp
from matplotlib.patches import Rectangle
from mpl_rdkit_utils import DrawMolToMPL
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
smi = "C[C@H](C(=O)OC(C)C)N[P@@](=O)(OC[C@H]1[C@@H]" + \
      "([C@]([C@H](O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3"
mol = Chem.MolFromSmiles(smi)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
DrawMolToMPL(mol, fig, ax)
fig.show()
```

```python
def get_smiles(queries):
    for q in queries:
        c = pcp.get_compounds(q, "name")[0]
        yield c.isomeric_smiles
```

```python
cnames = [
    "Aspirin",
    "Ibuprofen",
    "Acetaminophen",
    "Ethenzamide",
    "Loxoprofen",
    "Caffeine",
]
mols = [Chem.MolFromSmiles(smi) for smi in get_smiles(cnames)]
```

```python
fig, ax = plt.subplots(2, 3, figsize=(10, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a)
    a.set_title(cname)
fig.show()
fig.savefig("out_examples/out.pdf", bbox_inches="tight")
```

```python
fig, ax = plt.subplots(2, 3, figsize=(9, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a, style="golden")
    a.set_title(cname)
    a.set_axis_on()
    a.xaxis.set_visible(False)
    a.yaxis.set_visible(False)
fig.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(9, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a, style="none")
    a.set_title(cname)
    a.set_axis_on()
    a.xaxis.set_visible(False)
    a.yaxis.set_visible(False)
fig.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(9, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a, style="tight")
    a.set_title(cname)
    a.set_axis_on()
    a.xaxis.set_visible(False)
    a.yaxis.set_visible(False)
fig.tight_layout()
fig.show()
```

```python
smi = "C[C@H](C(=O)OC(C)C)N[P@@](=O)(OC[C@H]1[C@@H]" + \
      "([C@]([C@H](O1)N2C=CC(=O)NC2=O)(C)F)O)OC3=CC=CC=C3"
mol = Chem.MolFromSmiles(smi)
```

```python
fig, ax = plt.subplots()
(x0, x1), (y0, y1) = DrawMolToMPL(mol, fig, ax)
r = Rectangle(xy=(x0, y0), width=x1 - x0, height=y1 - y0, fill=False)
ax.add_patch(r)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_axis_on()
fig.show()
```
