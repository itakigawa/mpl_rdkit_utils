# DrawMolToMPL Example

```python
# !pip install pubchempy
import matplotlib.pyplot as plt
import pubchempy as pcp
from mpl_rdkit_utils import DrawMolToMPL
from rdkit import Chem
```

```python
def get_smiles(queries):
    for q in queries:
        c = pcp.get_compounds(q, "name")[0]
        yield c.isomeric_smiles
```

```python
cnames = [
    "Aspirin",
    "Ibuprofen",
    "Acetaminophen",
    "Ethenzamide",
    "Loxoprofen",
    "Caffeine",
]
mols = [Chem.MolFromSmiles(smi) for smi in get_smiles(cnames)]
```

```python
fig, ax = plt.subplots(2, 3, figsize=(10, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a)
    a.set_title(cname)
fig.show()
fig.savefig("out.pdf")
```

```python
fig, ax = plt.subplots(2, 3, figsize=(9, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a, style="golden")
    a.set_title(cname)
    a.set_axis_on()
    a.xaxis.set_visible(False)
    a.yaxis.set_visible(False)
fig.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(9, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a, style="none")
    a.set_title(cname)
    a.set_axis_on()
    a.xaxis.set_visible(False)
    a.yaxis.set_visible(False)
fig.show()
```

```python
fig, ax = plt.subplots(2, 3, figsize=(9, 5))
for m, a, cname in zip(mols, ax.flatten(), cnames):
    DrawMolToMPL(m, fig, a, style="tight")
    a.set_title(cname)
    a.set_axis_on()
    a.xaxis.set_visible(False)
    a.yaxis.set_visible(False)
fig.tight_layout()
fig.show()
```
